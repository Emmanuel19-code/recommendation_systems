```python
import ast
```

```python
#converting object data into a string data
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies_df["new_genres"] = movies_df["genres"].apply(convert)
```

```python
movies_df.head()
```

```python
movies_df["new_keywords"] = movies_df["keywords"].apply(convert)
```

```python
movies_df.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies_df["new_cast"] = movies_df["cast"].apply(convert3)
```

```python
def fetchdirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
    return L
```

```python
def convertsentence(text):
    sentence = []
    doc = nlp(text)
    for token in doc:
        sentence.append(token)
    return sentence
```

```python
movies_df["new_overview"] = movies_df["overview"].apply(convertsentence)
```

```python
movies_df["tags"] = movies_df["new_overview"]+movies_df["new_genres"]+movies_df["new_cast"]+movies_df["new_crew"]+movies_df["new_keywords"]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
stem = PorterStemmer()
```

```python
def stemming(text):
    y = []
    
    for i in text.split():
        y.append(stem.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stemming)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
def recommendMovie(movie):
    movie_index = new_df[new_df["title"]==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[:10]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommendMovie("Spectre")
```
